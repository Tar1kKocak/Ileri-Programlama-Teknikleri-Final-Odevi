```python
import pandas as pd

X_i = [15,15,16,19,19,20,20,21,22,28,35,40,41,42,43,44,60,61,65]
C_1 = 16
C_2 = 22



for x in range(4):
    
    t=0
    r=0
    
    Distance1 = []
    Distance2 = []
    
    araci1 =[]
    araci2= []
    
    toplam1=0
    toplam2=0
    
    Nex_Centroid = []
    
    Nearest_Cluster=[]


    for i in X_i:
        yardimci_a=0
        yardimci_b=0
        a = abs(i-C_1)
        b = abs(i-C_2)
        Distance1.append(a)
        Distance2.append(b)
        if a>b:
            yardimci_b+=2
            Nearest_Cluster.append(yardimci_b)
            toplam2+=C_2+i
            t+=2
        elif b>a:
            yardimci_a+=1
            Nearest_Cluster.append(yardimci_a)
            toplam1+=C_1+i
            r+=2
        else:
            yardimci_b+=2
            Nearest_Cluster.append(yardimci_b)
            toplam2+=C_2+i
            t+=2

    araci1=toplam1/r
    araci2=toplam2/t
   
    for j in Nearest_Cluster:
        if j == 1:
            Nex_Centroid.append(araci1)
        else:
            Nex_Centroid.append(araci2)
     

    a = {
         "X_i":X_i,
         "C_1":round(C_1, 2),
         "C_2":round(C_2, 2),
         "Distance 1":Distance1,
         "Distance 2":Distance2,
         "Nearest Cluster":Nearest_Cluster,
         "New Centroid":Nex_Centroid
        }
    target = "satisfied"
    exclude = []

    df = pd.DataFrame(a)
    
    C_1 = araci1
    C_2 = araci2
    print(df)
```
